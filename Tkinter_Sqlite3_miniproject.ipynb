```python
def showrecord():
    import sqlite3 as mysql 
    from  tkinter import ttk
    con=mysql.connect('test.db')
    cur=con.cursor()
    s="select * from Student"
    cur.execute(s)
    row=cur.fetchall()
    
    
    ws  = Tk()
    ws.title('Student Information')
    ws.geometry('500x500')
    ws['bg'] = '#AC99F2'
    f = Frame(ws)
    f.pack()
    cols=('roll_no','name','age','course','gender','subject')
    ft = ttk.Treeview(f,columns=cols,show='headings')
    ft.heading('roll_no',text='Roll No')
    ft.heading('name',text='Name')
    ft.heading('age',text='Age')
    ft.heading('course',text='Course')
    ft.heading('gender',text='Gender')
    ft.heading('subject',text='Subject')
    for r in row:
        a=r[0]     #fetch rollno
        b=r[1]     #fetch name
        c=r[2]     #fetch age
        d=r[3]     #fetch subject
        e=r[4]     #fetch gender
        f=r[5]
        ft.insert(parent='',index='end',values=(a,b,c,d,e,f))
    ft.pack()
    ws.mainloop()
```
